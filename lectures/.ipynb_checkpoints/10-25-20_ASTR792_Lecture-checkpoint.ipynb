```python
# Let's look at temperature dependencies of some common collisional rates
# We will start with the de-excitation rate (Draine 2.27)

T = np.logspace(1,4,100) #Choose a range of temperatures

sigma_de = 8.69e-8 /np.sqrt(T/1.0e4)

pylab.figure(1,figsize=(10,7))
pylab.plot(np.log10(T), sigma_de)
pylab.ylabel("$\\langle \sigma v \\rangle$",fontsize=30)
pylab.xlabel("$ log(T$ [K]$)$",fontsize=25);

```

```python
#Using Astropy units

#Astropy recognizes commonly-used physics and astronomy units, 
#and will do simple arithmetical operations with them. 
print(u.km/u.s)

#You can also define variables for compound units
new_unit = u.kg * u.km/u.s **2

#Units can be quickly standardized to common systems like cgs and SI
print(new_unit.si)
print(new_unit.cgs)
```

```python
#If you end up with a compound unit, you can also try to simplify it
print(new_unit.compose())

#Since this typically returns an array of several possibilities, 
#you can specify a specific index to access the one you wish
print(new_unit.compose()[0])
```

```python
#Most commonly, we will work with units + values, which are called quantities
#When you are dealing with a quantity, you can access the value and the units separately
quantity = 4.05e9 * u.Hz

print(quantity.value)
print(quantity.unit)
```
