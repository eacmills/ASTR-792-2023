# Radiative Trapping

So far, in an effort to fully describe the excitation of molecular gas, we have considered:

* the internal structure of the molecule (Einstein A values and other probabilities for transitions between energy levels set by quantum considerations related to the molecular geometry)
* the transfer of energy due to the rate of physical collisions with other particles (including cosmic rays)
* the transfer of energy from a background continuum radiation field


However, there is one other key consideration needed to make a complete characterization of molecular excitation. This is the transfer of energy from molecule A (as it emits a photon to go from a state with $E_u$ to $E_l$) to molecule B, which may be able to use that photon itself to go from $E_l$ to $E_u$. 

This process is known as radiative trapping, and it becomes important when the gas becomes optically thick to line emission (e.g., there is a very short mean free path for a photon of a given energy to be absorbed by a molecule in a low-energy state that can absorb that photon and become excited). 

Putting this together with all of the previous considerations we have discussed we finally arrive at the real deal: this is the basic physics that underlies observational ISM research which takes observed line intensities and uses them to determine the physical properties of the gas: its density, temperature, and column density. 

This is also not a simple problem, as it is non-local: one cannot consider a single molecule at a time, but has to consider the impact of molecules in one part of a cloud on the excitation of molecules in another part of the cloud. As we will see, it thus depends on the geometry of the gas cloud as well as its velocity structure, as a sufficient Doppler shifting of the emitted photons will change their frequency enough to avoid reabsorption. 

## Escape Probability Approximation

As is often the case, we will start by backing away from the full complexity of this problem, and making an approximation that greatly simplifies the math. 

Here, we begin by assuming that in many cases the problem is *not* as non-local as it seems, and we can begin to effectively decouple the excitation of a single molecule from the impact of molecules elsewhere in the cloud. We do this by determining, based on the geometry and velocity structure of the cloud, the likelihood $\beta_nu$ that a photon emitted at a frequency $\nu$ and a given location $r$ will be able to escape the cloud. This will be a function of its optical depth $\tau_\nu$ along all the possible paths with directions $\hat{n}$ that it could take:

$\beta_\nu (r) = \displaystyle \int \frac{d\Omega}{4\pi} e^{-\tau_\nu(r,\hat{n})}$

This is typically averaged over the normalized line profile $\phi_\nu$ (note that many basic radiative transfer codes just adopt a simplified rectangular line profile, while more sophisticated codes include more complicated line shapes):


$\langle \beta_\nu(r)\rangle = \displaystyle \int \phi_\nu \beta_\nu(r) d\nu$

We then construct a binary scenario which assumes that, based on this escape probability, either a photon is effectively absorbed by a molecule so close by that it can be assumed to be absorbed exactly where it was emitted (this is called the 'on the spot' approximation), or it leaves the cloud entirely. 

To further eliminate the importance of the location within the cloud on a molecule's excitation, we will also assume that the excitation is uniform across the entire cloud. In this case, we can write the local equation of radiative transfer at a point $r$ as:



$I_\nu = \displaystyle I_\nu(0)e^{-\tau_\nu} + B_\nu(T_{ex})(1-e^{-\tau_\nu})$

where as we recall, $B_\nu$ is the source function $j_\nu / \kappa_\nu$, and per our assumptions above is the same at any point in the cloud. 

To determine the level populations and so the overall excitation state of the gas, we return to our formula for the rate of change of the population of a given energy level $E_u$:

$\displaystyle \frac{dn_u}{dt} = n_l ( B_{lu} u_\nu +k_{lu} n_c) - n_u (A_{ul}+B_{ul}u_\nu+k_{ul} n_c)$ 

Here the $k$'s are the collisional rate coefficients, $u_\nu$ is the energy density of the radiation field, and $n_c$ is the number density of the collision partner (typically H$_2$ for molecular gas). Draine rewrites this in terms of the photon occupation number $n_\gamma$, removing both the dependence on the Einstein B and on $u_\nu$ and adding a dependence on the level degeneracy $g$ to get:

$\displaystyle \frac{dn_u}{dt} = n_l \left(n_c k_{lu}  + n_\gamma \frac{g_u}{g_l} A_{ul}\right)  - n_u (n_c k_{ul} + A_{ul}+n_\gamma A_{ul})$ 

We now apply the definition of $n_\gamma$, noting that $e^{-\tau_\nu}$ is just the escape probability $\beta_\nu$:

$n_\gamma(\nu) = \displaystyle \langle \beta \rangle n_\gamma^{(0)} + \frac{1-\langle\beta\rangle}{(n_l g_u / n_u g_l) -1} $

where

$n_\gamma^{(0)} \equiv \displaystyle \frac{c^2}{2 h \nu^3} I_\nu(0)$

Putting all of these pieces back together and collecting terms we end up with:


$\displaystyle \frac{dn_u}{dt} = n_c k_{lu} n_l  - n_c k_{ul}n_u  - \langle\beta\rangle A_{ul} n_u +  n_l\frac{g_u}{g_l}\langle\beta\rangle A_{ul}n_\gamma^{(0)} \left( 1- \frac{n_u g_l}{n_l g_u}\right)$ 

And this monster is the escape probability approximation. In order, these terms correspond to (1) collisional excitation (2) collisional de-excitation (3) spontaneous emission and (4) absorption and stimulated emission. 

As long as we have an expression for $\beta$, we can use this equation to solve for the level populations that result from a full consideration of all collisional processes *and* all radiative processes (absorption, stimulated and spontaneous emission) that occur due to photons from the background radiation field and photons that originate within the cloud itself. 

We can now also write a more accurate expression for the critical density of a transition (the density required for thermalization) that considers radiative trapping: 

$n_{crit,u} = \displaystyle \sum_c \left( \frac{\sum_{l<u}\langle\beta_{ul}\rangle A_{ul}}{\sum_{l<u}k_{ul}(c)} \right)$

This is a complete expression for this quantity that includes summing over all collision partners c, and all levels $l<u$ that have transitions connecting to $u$. 


## Geometry: Spherical cloud

We now consider how to determine a geometry-specific approximation for $\beta$. As is so often the case in physics, this takes the form of a spherical cow (or rather, cloud).  

If we consider a uniform-density static spherical cloud:

$\beta_{cloud} = \displaystyle \frac{3}{4 \pi R^3} \int_0^R \langle \beta(r) \rangle 4 \pi r^2 dr$

Other commonly assumed geometries that one may encounter include a homogeneous slab, a static shell, and expanding shells and spheres. 

Using the equation for optical depth:

$\tau_\nu = \displaystyle \frac{g_u}{g_l} \frac{A_{ul}}{8\pi} \lambda^2_{ul} \phi_\nu \int n_l \left( 1- \frac{n_u g_l}{n_l g_u}\right) ds$

We define the optical depth at line center going from the center of the cloud ($s=0$) to its surface ($s=R$) to be:

$\tau_0 \equiv \displaystyle \frac{g_u}{g_l} \frac{A_{ul}\lambda_{ul}^3}{4(2\pi)^{3/2} \sigma_V} n_l R  \left( 1- \frac{n_u g_l}{n_l g_u}\right)$

Here $\sigma_V$ is the 1D velocity dispersion, assuming a Gaussian line profile. (I am not going through the full maths here, but note the not-unexpected similarity to the last term of the equation describing the escape probability approximation, which corresponds to absorption and stimulated emission, e.g., processes that would prevent an emitted photon from escaping the cloud.)

$\beta_{cloud}$ can be well-fit by the function

$\beta_{cloud} \simeq \displaystyle \frac{1}{1+0.5 \tau_0}$

We can understand this result (this is the fraction of escaping photons averaged over the whole cloud) as corresponding to the fraction of the cloud's mass (e.g., the number fraction of equal-mass molecules in the cloud) that lie within a distance from the surface corresponding to $\tau \sim 2/3$ -- or the "photosphere" of the cloud. 

Ultimately, the determination of a $\beta$ is an iterative problem: guess a $\beta$, solve for the level populations, use these to calculate $\tau$ and a new value of $\beta$, repeat until a consistent result is arrived at. 

So yes, guess and check is a real thing that real scientists do. 

Generally, one has some constraints on the level populations directly from observations of the escaping photons that originate from a transition between some upper and lower energy levels. However, one generally cannot observe every allowed transition between every possible energy level. So, to determing the approximate the fraction of molecules in a given level of interest compared to all molecules in all levels, we turn to our old friend the partition function. 

The relative population in a rotational level $J$ for a linear molecule is given by:

$\displaystyle \frac{n(J)}{n} = \frac{(2J+1)\exp(-B_0 J(J+1)/ k_B T_{ex})}{\sum_J (2J+1)\exp(-B_0 J(J+1)/ k_B T_{ex})}$

The denominator of this expression is the partition function:

$\displaystyle Z \equiv \sum_J (2J+1)\exp(-B_0 J(J+1)/ k_B T_{ex}) \approx \left[1 + \left(\frac{k_B T_{ex}}{B_0}\right)^2 \right]^{0.5}$

## Velocity Field: Large Velocity Gradient

The simple escape fraction approximation assumed that all of the molecules in the cloud are at rest (though of course they are still undergoing collisions). A more realistic assumption is that the particles are actually moving, either with some uniform bulk motion, or randomly due to turbulence. As mentioned before, this has an important impact on the likelihood that a photon will be reabsorbed, as molecules in the cloud will "see" each other (and photons emitted by other molecules) as having some red- or blue-shift. 

This will change the optical depth $\tau$ as well as the resulting escape probability $\beta$. 

Two important cases are that of a uniform ('Hubble') flow like that in an expanding sphere, and that of a uniformly turbulent medium.

The first is typically referred to as the "large velocity gradient" or LVG approximation. For a sphere expanding at every point with a rate $|dv/dr|$, the escape probability can be written as

$\langle \beta \rangle \displaystyle = \frac{1-e^{-\tau_{LVG}}}{\tau_{LVG}}$

Here, the optical depth for the LVG approximation is just

$\tau_{LVG} \equiv \displaystyle \frac{\lambda_{ul}\int \kappa_\nu d\nu}{|dv/dr|}$

which can be understood as the total optical depth across the cloud for a photon at any frequency that could be absorbed by a nearby molecule:

$\nu$ in $\nu_0[1\pm(R/c)(dv/dr)]$

For turbulent clouds, one just adapts the earlier expressions for $\beta$ and $\tau$ for a uniform, static cloud to effectively encompass a larger line width corresponding to the turbulent broadening. 

One can take the specific case of line broadening in a self-gravitating cloud, defined by the virial theorem. Here, 

$\sigma_V^2 = \displaystyle \frac{GM}{5R}$

This value can be plugged in to the expression we previously saw for $\tau_0$:

$\tau_0 \equiv \displaystyle \frac{g_u}{g_l} \frac{A_{ul}\lambda_{ul}^3}{8 \pi} \left( \frac{5}{2\pi G}\right)^{0.5} \frac{n_l R^{3/2}}{M^{1/2}}  \left( 1- \frac{n_u g_l}{n_l g_u}\right)$

Finally, relating this back to observable quantities, we can express the total cloud luminosity in a given spectral line as:

$L_{ul} = \displaystyle \int 4\pi r^2 dr n_u A_{ul} h\nu_{ul} \langle \beta \rangle_{cloud} \simeq \frac{4\pi}{3}R^3 n_u A_{ul}h \nu_{ul} \frac{1}{1+0.5 \tau_0}$

This then relates the observed brightness of a spectral line to the number of molecules in a level with energy $E_u$, allowing us to go from an observed line intensity to a determination of the level populations $n_u$. 

While this exact expression describes a (turbulent) sphere, one can also write expressions for other common geometries (e.g., shells, cylinders, or uniform slabs).
