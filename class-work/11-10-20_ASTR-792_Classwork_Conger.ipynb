```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from astropy import units as u
```

The Boltzmann equation can be used to measure excitation temperature using observations multiple transitions of a molecule (see the September 08 lecture) :

$\displaystyle \frac{N_u}{N_l} = \frac{g_u}{g_l} \exp\left(\frac{E_l - E_u}{k_B T_{ex}}\right)$

(1) Rewrite the Boltzmann equation as a function of $T_{ex}$, log$_{10}(N_u/g_u)$, and log$_{10}(N_l/g_l)$, and put this into the form of a slope of a line, e.g., $m = \left(\frac{y_2-y_1}{x_2-x_1}\right)$

(2) Use this relation to plot the following data set of NH$_3$ observations as a function of $E$.



(3) Fit a slope to these data and use this to determine the excitation temperature (Note that for NH$_3$, T$_{ex}$ measured in this way is a good approximation for the true kinetic temperature $T_{kin}$).

## Problem One

With my nonexistent algebra prowess, I rearrange the eqaution to yield the following:

$ {T_{ex}(\ln(N_u g_l) - \ln(N_l g_u))} = \frac{E_l - E_u}{k_B} $

and

$ k_B T_{ex} = \frac{E_l - E_u}{\ln(N_u g_l) - \ln(N_l g_u)} $

## Problem Two & Three

I use...

$\log(N_u g_l) - \log(N_l g_u) = -\frac{1}{T_{ex}} \frac{E_u - E_l}{k_B}$

That is, $1/T_{ex}$ will be the slope.

```python
E_u = np.array([852.8,1036.4,1237.6,1456.4,1692.7])
E_l = np.array([686.8,852.8,1036.4,1237.6,1456.4]) 
delE = np.abs(E_l-E_u)

N_u = np.array([6.1e13,1.5e13,9.0e12,1.4e13,3.5e12])
N_l = np.array([4.8e13,6.1e13,1.5e13,9.0e12,1.4e13])

#for 9-->8 transition, for instance, delE[0], N_u[0], and N_l[0]
```

```python
#rotational degeneracy numbers 2J+1

J = np.array([8,9,10,11,12,13])
gJ_u = 2*J[1:6] + 3                                 #I add a 2 to the degeneracy to account for K
gJ_l = 2*J[0:5] + 3
```

```python
y = np.log10(N_u * gJ_l) - np.log10(N_l*gJ_u)
x = delE #E_l - E_u / k_B
m,b = np.polyfit(x,y,1) #assumes linear relationship
plt.scatter(x,y)
plt.plot(x,m*x+b)
print(1/m * u.K)

plt.xlabel('Energy',fontsize=15)
plt.ylabel('Log hoopla',fontsize=15)
```
