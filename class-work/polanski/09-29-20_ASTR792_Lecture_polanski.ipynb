# Rate Balancing

###  Ionization Rate

The rate of photoionizations per time is given as:

$\zeta_{pi} = \displaystyle \int_{\nu_1}^{\infty} \sigma_{pe}(\nu)\; c \frac{u_\nu}{h\nu} d\nu$

We can then write the ionization rate (photoionizations per volume per time) for species $X$ as:

$\displaystyle \frac{dn_e}{dt} = n_X \zeta_{pi}$

### Recombination Rate

As we recall, the thermal rate coefficient (the likelihood that a free electron will be captured into state $nl$, resulting in recombination and the emission of a photon) is:


$\alpha_{nl} = \displaystyle \left(\frac{8kT}{\pi m_e}\right)^{1/2} \int_0^\infty \sigma_{rr,nl}(E) \frac{E}{kT} e^{-E/kT} \frac{dE}{kT}$


Like other rate coefficients, this has units of volume per time (e.g., cm$^{3}$ s$^{-1}$).
For Case B recombination, a functional approximation for the recombination coefficient is 

$\alpha_B = \displaystyle 2.54 \times10^{-13} Z^2 \left(\frac{T_4}{Z^2}\right)^{-0.8163 -0.0208 \ln (T_4/Z^2)}$

We can then write the recombination rate (recombinations per volume per time) for species $X$ with charge $Z$ as: 

$\displaystyle \frac{dn_e}{dt} = - n_{X^+} n_e \alpha_B$

The equilibrium condition (rate of recombinations equalling the rate of photoionizations in some volume) is then achieved when $\displaystyle \frac{dn_e}{dt} =0$ or:

$n_X\zeta_{pi} = \displaystyle n_{X^+} n_e \alpha_B$

# Stromgren Sphere


We will use this expression for equilibrium to calculate the size of a pure-hydrogen HII region around an ionization source (e.g., a massive star). 

We will define the size of the region that can be ionized as the region within a radius $R$. Within this volume, the hydrogen is assumed to be totally ionized, with the total number of recombinations balancing the total number of ionizations, and ionizing photons unable to penetrate beyond this volume. We can then write

$\displaystyle \int_0^R n_H  \int_{\nu_1}^{\infty} \sigma_{pe}(\nu)\; c \frac{u_\nu}{h\nu} 4\pi r^2 dr d\nu = \int_0^R n_{H^+} n_e \alpha_B 4\pi r^2 dr$

The right hand side of this expression is easy to integrate, and we can rewrite it as 

$\displaystyle \int_0^R n_H  \int_{\nu_1}^{\infty} \sigma_{pe}(\nu)\; c \frac{u_\nu}{h\nu} 4\pi r^2 dr d\nu = \frac{4}{3}\pi R^3 n_{H^+} n_e \alpha_B $


The left-hand side will take a little more thought. We will first look at the term $n_H \int \sigma_{pe}(\nu)\; c d\nu $. The quantity inside the integral is a rate coefficient, with units of volume per time, and $n_H$ has units of per volume. In essence, this is a probability (per time) that, given a photon of the correct energy, it will result in an ionization. Because we are considering the probability that this will occur somewhere in the entire HII region, we can safely set this probability to just be 1. (Note that as this is still technically a probability PER TIME, we will want to keep track of those units). This leaves us with:

$\displaystyle \int_0^R \int_{\nu_1}^{\infty} \frac{u_\nu}{h\nu} 4\pi r^2 dr d\nu = \frac{4}{3}\pi R^3 n_{H^+} n_e \alpha_B $


We can define the number of emitted photons capable of ionizing hydrogen (ionization threshold $I = h\nu_I = 13.6$ eV for hydrogen) as $N_0$ where:

$\displaystyle  \frac{N_0}{V} = \int_{\nu_I}^\infty \frac{u_\nu}{h\nu} d\nu $


The right hand side of this expression is the integral of the radiation energy density divided by the typical energy per photon (recall that $u_\nu = \displaystyle \frac{1}{c} \int I_\nu d\Omega $ with units of energy per volume per frequency), and so gives us the total number of photons per some volume. We will go ahead and pick up those leftover units of per time from the first part of this integral, and define a quantity $Q_0$ which is now not the number of ionizing photons but the rate at which ionizing photons are emitted per time. 

We can use this to rewrite our rate-balancing equation as 

$\displaystyle \frac{Q_0}{V} \int_r^R 4\pi r^2 dr = Q_0 =  \frac{4}{3}\pi R^3 n_{H^+} n_e \alpha_B $

Assuming again that this is a pure-Hydrogen nebula, and furthermore that the hydrogen is fully ionized, we can say $n_H = n_e = n_{H^+}$ and solving for $R$ we can define the Stromgren radius of a pure-hydrogen HII region as:


$\displaystyle R_s \equiv \left( \frac{3 Q_0}{4 \pi n_H^2 \alpha_B}\right)^{1/3}$

In practice, as an observer, one often wants to use this relation to solve for something that is less directly observable than the radius of the nebula, like the ionizing flux $Q_0$, which can tell you about the type of ionizing star, or the electron density $n_e$, if you already know the stellar type (say, from spectroscopy of the central star).

# Time Scales

When one discusses rates of processes, or the amount of change in a quantity as a function of time, one inevitably also wants to define timescales on which these quantities undergo some meaningful change. In this case, it could be the time it takes to go from neutral gas to fully ionized gas when a star first "switches on", or the time that it takes for gas to become neutral again if a star is moving through the ISM and the gas is no longer within its Stromgren radius. 

## Ionization time scale

Most simply, the ionization time scale is just the inverse of the ionization rate (ionizations per time or $\displaystyle \frac{1}{ \zeta_{pi}}$)

However, it is generally more useful to parametrize this in terms of the rate of emission of ionizing photons $Q_0$. Here the time scale is defined by total number of neutral atoms which must be ionized divided by this photon production rate:

$\displaystyle \tau_i = \frac{(4/3)\pi R_S^3 n_H}{Q_0}$


## Recombination time scale

In the same way, we can define a recombination time scale as the inverse of the recombination rate (note that this is recombinations per unit time, *not* recombinations per unit time per unit volume):

$\displaystyle \tau_r = \frac{1}{n_{H+} \alpha_B}$

Note that for a nebula in ionization equilibrium (which is generally a good assumption for HII regions), these two time scales are identical. This is similar to what we see when we consider objects (like stars) in a dynamical equilibrium: the time scale for contraction (the free-fall time) is the same as the time scale for expansions (the sound-crossing time). Speaking of which...

## Dynamic time scale (Sound-crossing time)


It is often useful to compare the time scales of all the different processes that can affect the structure of a region. While HII regions can typically be assumed to be in ionization equilibrium, it is not generally a good assumption that they are in pressure equilibrium. 

The characteristic time scale for changes in the nebula structure due to pressure waves is the sound-crossing time: the time it takes information traveling at the sound speed to cross a distance equal to twice the radius of the nebula:

$\displaystyle \tau_{sc} = \frac{2 R_S}{c_s}$

Here $c_s$ is the isothermal sound speed, defined as 

$\displaystyle c_s = \left( \frac{2kT}{m_H}\right)^{1/2}$

Next week, we will get the chance to compare this to another time scale of interest: the cooling time, which will tell us whether or not we can assume that a nebula is in thermal equilibrium (spoiler: we used the isothermal sound speed, so that's a good assumption), with a constant electron temperature across the nebula. 


# Metals

Of course, just like our assumption that the hydrogen in an HII region is fully-ionized, our assumption that the HII region is fully H is also faulty (and as we will later see, this has important consequences for the temperature structure of HII regions!). For now, we will concentrate on what this means for the ionization structure. 

Different atoms have very different ionization thresholds, so (depending on the hardness of the spectrum of the ionizing source), the radius of the region in which these atoms are fully (multiply) ionized may be smaller than the HII stromgren sphere (or larger in the case of atoms like carbon, which has a lower ionization threshold and so can be ionized by radiation with $h\nu < 13.6$ eV that escapes the HII region.  

One of the most typical to consider is He, which has an ionization energy almost twice that of Hydrogen ($h\nu = 24.6$ eV). When both H and He are present, some fraction $y$ of the $h\nu > 24.6$ eV photons which would have been responsible for ionizing hydrogen now go toward ionizing helium. In return however, some fraction $z$ of the photons emitted by recombining helium will be sufficiently energetic to ionize hydrogen.

This means that we have to rewrite our rate-balancing equations. For hydrogen, where we had 

$\displaystyle N_{H^+}n_e \alpha_{B,H} = Q_0$  (note that $N_{H^+}$ is a number, not a number density)

now we must write:
$\displaystyle N_{H^+}n_e \alpha_{B,H} = \left[Q_0-(1-y)Q_1\right] + N_{He^+} n_e \left[z\alpha_{B,He} + y\alpha_{1S}\right] $ 

Here $\alpha_{1S}$ is the recombination rate to the Helium ground state; note that these photons can ionize He or H. Breaking it down term-by-term, the left hand side is still the number of Case B recombinations, while the right hand side is the number of H-ionizing photons due to (1) the H-ionizing photons emitted by the star, minus those which are stolen by helium (2) the photons emitted from helium recombinations to $n=2$ that are energetic enough to ionize hydrogen (3) the photons emitted from helium recombinations to $n=1$ that are energetic enough to ionize hydrogen.


Similarly, for Helium, we can write the condition for balance as

 

$\displaystyle N_{He^+}n_e \left[\alpha_{B,He} + y\alpha_{1S,He}\right] = (1-y) Q_1$  

Breaking this one down term-by-term, the left hand side is the number of recombinations that do *not* result in photons that could ionize helium. This includes recombinations to $n=2$ as well as those recombinations to $n=1$ that result in a photon that (while it could ionize helium) instead goes into the ionization of hydrogen. The right hand side is the number of helium-ionizing photons (minus those stolen by hydrogen).

Putting these together, with some reasonable assumptions for $y$ and $z$ ($y=0.2$, $z=0.8$) and presumably some assumptions for the recombination coefficients, which Draine does not give any explicit numbers for, we can then say

$\displaystyle \frac{N_{He^+}}{N_{H^+}} \sim \frac{0.68 (Q_1/Q_0)}{1-0.17(Q_1/Q_0)}$

Helium will be fully ionized in the entire volume of the HII region when 

$\displaystyle \frac{N_{He^+}}{N_{H^+}} = \frac{n_{He}}{n_{H}} = 0.096$.

This requires $Q_1/Q_0 \gtrsim 0.15$

## Example: Draine 15.1(b)

Here we consider the size of the region in a nebula in which He can be assumed to be fully ionized. 

We know that $Q_0 = 10^{48.06}$ s$^{-1}$, $Q_1 = 0.0145 Q_0$ (the rate of emission of He-ionizing photons), $n_{He} = 0.1$, $n_H = 10^3$ cm$^{-3}$, and T = $10^4$ K. 

We want to know the ratio $R_{HeII}/R_{HII}$. 


The first thing we can do compare $\displaystyle \frac{N_{He^+}}{N_{H^+}}$ with $\displaystyle \frac{n_{He}}{n_{H}}$

We find that $\displaystyle \frac{N_{He^+}}{N_{H^+}} \sim \frac{0.68 (0.0145)}{1-0.17(0.0145)} \sim 0.0098$

This is less than $\displaystyle \frac{n_{He}}{n_{H}}$, and so we expect the radius of the HeII region to be smaller than the radius of the HII region. To determine by how much, we will generalize Draine's Equation 15.36 to say

$\frac{R_{HeII}}{R_{HII}} \sim \left(\frac{n_H}{n_{He}}\frac{N_{He^+}}{N_{H^+}}\right)^{1/3}\sim$ 0.2

```python
import numpy as np
import astropy.units as u
import astropy.constants as const

#ionization timescale

Q0 = (10**48.06) * (1/u.s)
#Q1 = 0.0145*Q_0

n_H = 1e3 * (u.cm**-3)
T = 1e4 * u.K
a_B = 2.54e-13 * (u.cm**3 / u.s)

R_s = (3*Q0 / (4*np.pi*(n_H**2)*a_B))**(1/3)

t_i = (4/3)*np.pi*(R_s**3)*n_H / Q0
print(f"Ionization Timescale = {t_i.to(u.year):0.4} ")

#recombo timescale

t_r = 1 / (n_H * a_B)
print(f"Recombination Timescale = {t_r.to(u.year):0.4}, equal to Ionization Timescale, as expected.")

#sound crossing time

mh = 1.674e-27 * u.kg
cs = np.sqrt(2*const.k_B*T / mh)
t_s = 2*R_s / cs
print(f"Sound Crossing Time = {t_s.to(u.year):0.6}")
```
