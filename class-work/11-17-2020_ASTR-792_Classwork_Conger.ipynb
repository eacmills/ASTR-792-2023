This image is a "mystery" spectrum taken with the ALMA observatory toward the center of a nearby starburst galaxy (NGC 253). You can assume that the gas in this environment is dense and warm ($\sim300$ K).



A table of measured line frequencies is below. The rest velocity of the gas in this part of the galaxy is $\sim200\pm10$ km/s.



Using this information, identify as many of these lines as you can!

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from astropy import units as u
```

```python
meas_nu = np.array([354.2535,354.4666,354.8141,355.0589,355.2058,355.3361,355.8505,356.0264,356.4877,356.9317,
                   357.0044,357.1530,357.3464,357.4391]) *1e9
```

```python
#using the Doppler redshift equation (rearranged to solve for the rest frequency),

v = (200 * u.km).to(u.m)
dv = (10 * u.km).to(u.m)
c = 2.998e8 * u.m / u.s

#the rest frequencies, calculated using the average veclocity, the upper bound velocity, and the lower bound velocity:

rest_nu = meas_nu/(1-v.value/c.value)
rest_nu_upper = meas_nu/(1-(v.value+dv.value)/c.value)
rest_nu_lower = meas_nu/(1-(v.value-dv.value)/c.value)
```

```python
print(rest_nu_upper[0]/1e9)
print(rest_nu_lower[0]/1e9)
print('-------------')
print(rest_nu[0]/1e9)
```
